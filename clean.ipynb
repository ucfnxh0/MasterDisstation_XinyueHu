```python
#merge
import pandas as pd
from math import radians, cos, sin, asin, sqrt
from tqdm import tqdm



# ---------- Haversine ----------
def haversine(lon1, lat1, lon2, lat2):
    R = 6371  
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon, dlat = lon2 - lon1, lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    return 2 * R * asin(sqrt(a))

# ---------- century_year ----------
people_cleaned['century_year'] = ((people_cleaned['active_year'] // 100) + 1) * 100
```

```python
# ---------- nearest city match ----------
radius_km = 50
matched_records = []

for _, person in tqdm(people_cleaned.iterrows(), total=len(people_cleaned)):
    plat, plon, ayear = person['birth_lat'], person['birth_long'], person['active_year']
    century_year = person['century_year'] 
    candidates = city_cleaned[city_cleaned['year'] == century_year].copy()

    if candidates.empty:
        continue

    candidates['distance'] = candidates.apply(
        lambda row: haversine(plon, plat, row['longitude'], row['latitude']), axis=1
    )

    nearest = candidates.loc[candidates['distance'].idxmin()]
    if (nearest['distance'] <= radius_km):
        combined = pd.concat([person, nearest])
        matched_records.append(combined)

matched_df = pd.DataFrame(matched_records)

matched_df.to_csv('matched_people_with_city.csv', index=False)
```

```python
import numpy as np
import pandas as pd
from scipy.sparse import block_diag
from spreg import ML_Lag
from libpysal.weights import KNN

y_var = 'log_pop'
X_vars = ['diversity','capital','commune','bishop','num_people_divide','university']

cities_order = list(cities_geo.sort_values('city')['city'])
years = sorted(panel_balanced['year'].unique())
N = len(cities_order)
T = len(years)

def stack_by_year(df, col):
    return np.vstack([
        df.loc[df['year']==y].set_index('city').reindex(cities_order)[[col]].values
        for y in years
    ])

def stack_X_by_year(df, cols):
    return np.vstack([
        df.loc[df['year']==y].set_index('city').reindex(cities_order)[cols].values
        for y in years
    ])

y_vec = stack_by_year(panel_balanced, y_var)
X_mat = stack_X_by_year(panel_balanced, X_vars)


from scipy.sparse import csr_matrix

W_blocks = [w.sparse.tocsr()] * T
W_big_sparse = block_diag(W_blocks, format='csr')  
```

```python
model = ML_Lag(
    y=y_vec,
    x=X_mat,
    w=W_big_sparse,
    slx_lags=1,         
    slx_vars='all', 
    constant=False,      
    name_y=y_var,
    name_x=X_vars
)
```

```python
import pandas as pd
import numpy as np


coefs = model.betas.flatten()
std_errs = model.std_err.flatten()
z_stats = model.z_stat
rho = model.rho
rho_z = z_stats[0][0]
rho_p = z_stats[0][1]
rho_se = abs(rho / rho_z) if rho_z != 0 else np.nan


try:
    varnames = model.name_x 
except:
    raise ValueError("Variable names cannot be obtained from the model")


rows = []

rows.append({
    'Variable': 'rho',
    'Coefficient': rho,
    'Std_Error': rho_se,
    'p_value': rho_p
})


for i, name in enumerate(varnames):
    pval = z_stats[i+1][1] if i+1 < len(z_stats) else np.nan
    rows.append({
        'Variable': name,
        'Coefficient': coefs[i],
        'Std_Error': std_errs[i],
        'p_value': pval
    })

coef_table = pd.DataFrame(rows)


print("SDM:")
print(coef_table)

coef_table.to_csv("sdm_coef_pvalues.csv", index=False, encoding="utf-8-sig")
print("success: sdm_coef_pvalues.csv")



```

```python
#for spatial spillover heatmap,influence value
from libpysal.weights import WSP
import numpy as np
import pandas as pd

def add_spatial_lags(df, w, vars_to_lag, id_order, years):
    df_new = df.copy()
    for var in vars_to_lag:
        lagged_vals = []
        for year in years:
            df_year = df_new[df_new['year'] == year].set_index('city').reindex(id_order)
            vals = df_year[var].values.reshape(-1, 1)
            W_mat = w.sparse.tocsr()
            lagged = W_mat @ vals
            lagged_vals.extend(lagged.flatten())
        df_new[f'W_{var}'] = lagged_vals
    return df_new

vars_to_lag = ['diversity', 'num_people_divide']
id_order = list(cities_geo.sort_values('city')['city'])
years = sorted(panel_balanced['year'].unique())

panel_balanced = add_spatial_lags(panel_balanced, w, vars_to_lag, id_order, years)

coef_table = pd.read_csv("sdm_coef_pvalues.csv", encoding="utf-8-sig")
theta_div = coef_table.loc[coef_table['Variable'] == 'W_diversity', 'Coefficient'].values[0]
theta_num = coef_table.loc[coef_table['Variable'] == 'W_num_people_divide', 'Coefficient'].values[0]

output_df = panel_balanced[['city', 'year', 'longitude', 'latitude', 'W_diversity', 'W_num_people_divide']].copy()
output_df['contribution_diversity'] = output_df['W_diversity'] * theta_div
output_df['contribution_num_people'] = output_df['W_num_people_divide'] * theta_num

mean_contrib = (
    output_df
    .groupby('city', as_index=False)
    .agg({
        'longitude': 'median',
        'latitude': 'median',
        'contribution_diversity': 'mean',
        'contribution_num_people': 'mean'
    })
)

mean_contrib.to_csv("spatial_contribution_average.csv", index=False, encoding="utf-8-sig")
print("success: spatial_contribution_average.csv")
```
